### Pandas - DataFrame and Series

```python
import pandas as pd
```

### Series

```python
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
```

```python
# Create a series from dictionary element
data = {
    "a" : 1,
    "b" : 2,
    "c" : 3,
    "d" : 4
}
series = pd.Series(data)
print(series)
```

```python
data = [1,2,3,4,5]
index = ["A", "B", "C", "D", "E"]
series = pd.Series(data, index = index)
print(series)
```

### DataFrame

```python
# Create a Data Frame from a list of Dictionaries
data = [
    {'Name' : 'Aadarsh', 'Age' : 21, 'City' : 'Gopalganj'},
    {'Name' : 'Aniket', 'Age' : 23, 'City' : 'Patna'},
    {'Name' : 'Ritesh', 'Age' : 23, 'City' : 'Chhapra'}
]
df = pd.DataFrame(data, index = [1,2,3])
print(df)
```
